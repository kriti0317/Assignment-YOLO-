```python
%matplotlib inline

import os
import cv2
import matplotlib.pyplot as plt

# Set paths
img_dir = "dataset/images/train"          # folder with your images
label_dir = "dataset/labels/train"        # folder with your YOLO label .txt files
class_names = [
    "dog", "person", "cat", "tv", "car", "meatballs", "marinara sauce",
    "tomato soup", "chicken noodle soup", "french onion soup", "chicken breast",
    "ribs", "pulled pork", "hamburger", "cavity", "bargraph", "cube", "circle",
    "cone", "graph", "text", "table", "figure", "box"
]

# Choose an image to visualize
img_file = "BAR CHART_page2.jpg"  # Replace with your image file name
img_path = os.path.join(img_dir, img_file)
label_path = os.path.join(label_dir, img_file.rsplit('.', 1)[0] + ".txt")

# Load image
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
h, w, _ = image.shape

# Draw boxes from YOLO label
if os.path.exists(label_path):
    with open(label_path, 'r') as f:
        for line in f:
            cls, x, y, bw, bh = map(float, line.strip().split())
            x1 = int((x - bw / 2) * w)
            y1 = int((y - bh / 2) * h)
            x2 = int((x + bw / 2) * w)
            y2 = int((y + bh / 2) * h)
            class_id = int(cls)
            label = class_names[class_id] if class_id < len(class_names) else str(class_id)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
else:
    print(f"No label file found for {img_file}")

# Show the image
plt.figure(figsize=(10, 8))
plt.imshow(image)
plt.axis("off")
plt.title("YOLO Labels Visualized")
plt.show()
```
